```python
# â•­â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â•®
# â”‚ 3. è°ƒç”¨è±†åŒ…ï¼ˆDoubaoï¼‰                                â”‚
# â”‚                                                      â”‚
# â”‚ æ ¹æ®ç«å±±æ–¹èˆŸæ–‡æ¡£ï¼š                                    â”‚
# â”‚ - base_url: https://ark.cn-beijing.volces.com/api/v3 â”‚
# â”‚ - model: ä½¿ç”¨å®˜æ–¹ Model IDï¼ˆå¦‚ doubao-seed-1.6ï¼‰     â”‚
# â”‚                                                      â”‚
# â”‚ æ–‡æ¡£å‚è€ƒï¼š                                           â”‚
# â”‚ https://www.volcengine.com/docs/82379/1399008        â”‚
# â•°â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â•¯

"""
è°ƒç”¨è±†åŒ…æ¨¡å‹ï¼ˆéœ€å…ˆåœ¨ç«å±±æ–¹èˆŸæ§åˆ¶å°å¼€é€šå¯¹åº”æ¨¡å‹ï¼‰

ğŸ“Œ ä½¿ç”¨å‰è¯·ç¡®ä¿ï¼š
  1. å·²åœ¨ç«å±±æ–¹èˆŸåˆ›å»º ARK_API_KEY
  2. å·²åœ¨ã€Œå¼€é€šç®¡ç†ã€ä¸­æ¿€æ´»ç›®æ ‡æ¨¡å‹

âœ… æ”¯æŒçš„ Model ID ç¤ºä¾‹ï¼ˆéœ€å…ˆå¼€é€šæ‰èƒ½ä½¿ç”¨ï¼‰ï¼š
  - doubao-seed-1.6          ï¼ˆå…¨èƒ½ç‰ˆï¼‰
  - doubao-seed-1.6-lite     ï¼ˆè½»é‡ç‰ˆï¼‰
  - doubao-1.5-pro-32k       ï¼ˆé•¿ä¸Šä¸‹æ–‡ï¼‰

ğŸ” æŸ¥çœ‹å¹¶ç®¡ç†ä½ å·²å¼€é€šçš„æ¨¡å‹ï¼š
  â†’ ç™»å½•ç«å±±æ–¹èˆŸæ§åˆ¶å°ï¼šhttps://console.volcengine.com/ark
  â†’ è¿›å…¥ã€æ¨¡å‹æœåŠ¡ã€‘â†’ã€å¼€é€šç®¡ç†ã€‘
  åœ¨æ­¤é¡µé¢å¯ï¼š
    â€¢ æœç´¢æ¨¡å‹å¹¶ç‚¹å‡»ã€Œå¼€é€šã€
    â€¢ æŸ¥çœ‹å½“å‰è´¦å·ä¸‹æ‰€æœ‰å·²å¼€é€šçš„æ¨¡å‹åŠå…¶ Model ID

ğŸ“š å®˜æ–¹å®Œæ•´æ¨¡å‹åˆ—è¡¨ï¼ˆå«èƒ½åŠ›è¯´æ˜ï¼‰ï¼š
  https://www.volcengine.com/docs/82379/1330310
"""

from openai import OpenAI

# åˆå§‹åŒ–ç«å±±æ–¹èˆŸå®¢æˆ·ç«¯
client = OpenAI(
    api_key=ARK_API_KEY,
    base_url="https://ark.cn-beijing.volces.com/api/v3"
)

def ask_doubao(prompt: str, model: str = "doubao-seed-1-8-251228") -> str:
    completion = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7,
    )
    return completion.choices[0].message.content

# æµ‹è¯•è°ƒç”¨
response = ask_doubao("ä½ å¥½ï¼è¯·ç”¨ä¸€å¥è¯ä»‹ç»ä½ è‡ªå·±ã€‚")
print("ğŸ¤– è±†åŒ…å›ç­”ï¼š")
print(response)
```
